```python
#import libreries
import numpy as np
import pandas as pd 
```

```python
# importing and Reading dataset

ds= pd.read_csv('cleaned_food_prices.csv')
ds = ds.loc[:, ~ds.columns.str.contains('^Unnamed')]                # removing unnamed column
ds
```

```python
ds.info()
```

```python
# splitting dataset into Independent and Dependent

x=ds[['Country', 'Global_Rank']]
y=ds[['Price_2017']]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
# 3. Create Column Transformer
ct = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ['Country']),  # only encode categorical
    remainder='passthrough'  # keep numeric as is
)
ct
```

```python
from sklearn.linear_model import LinearRegression
pipe = make_pipeline(ct, LinearRegression())
```

```python
from sklearn.metrics import r2_score  # <-- Required import
scores = []
for i in range(0, 101):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=i)
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
# finding best score
bestindex=np.argmax(scores)
scores[bestindex]
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1,random_state=bestindex)
pipe.fit(x_train, y_train)
```

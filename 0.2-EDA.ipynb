```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_csv('cleaned_food_prices.csv')
df
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]                # removing unnamed column
df
```

# Distribution of Prices

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Price_2017'], kde=True, color='skyblue')
plt.title("Distribution of Milk, Cheese, and Eggs Prices (2017)")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
 # Top 10 Most Expensive Countries
top10 = df.sort_values(by='Price_2017', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Price_2017', y='Country', data=top10, palette='Reds_r')
plt.title("Top 10 Countries by Food Price (2017)")
plt.xlabel("Price (USD)")
plt.ylabel("Country")
plt.show()
```

```python
# Bottom 10 Cheapest Countries

bottom10 = df.sort_values(by='Price_2017', ascending=True).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Price_2017', y='Country', data=bottom10, palette='Greens_r')
plt.title("Bottom 10 Countries by Food Price (2017)")
plt.xlabel("Price (USD)")
plt.ylabel("Country")
plt.show()
```

```python
# Price vs Global Rank
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Global_Rank', y='Price_2017', color='purple')
plt.title("Price vs Global Rank")
plt.xlabel("Global Rank")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

```python
# correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Price_2017', 'Global_Rank']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Boxplot to Detect Outliers
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, y='Price_2017')
plt.title("Boxplot of Prices (2017)")
plt.ylabel("Price (USD)")
plt.show()
```
